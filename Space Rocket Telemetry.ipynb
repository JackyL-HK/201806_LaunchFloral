```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)

import numpy as np

x = fhd1_s1[21:320]['x(km)']
y = fhd1_s1[21:320]['y(km)']
z = fhd1_s1[21:320]['z(km)']
x2 = ird6_s1[21:320]['x(km)']
y2 = ird6_s1[21:320]['y(km)']
z2 = ird6_s1[21:320]['z(km)']
time = fhd1_s1[21:320]['#Time']
relVel50 = [vel / 50 for vel in fhd1_s1[21:320]['relVel']]

trace = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    text=time,
    mode='lines+markers',
    line=dict(width=5),
    marker=dict(
        size=relVel50,
        color=relVel50,
        colorscale=[[0, 'rgb(0,0,255)'], [1, 'rgb(255,0,0)']]),
    name='Trace 1')
data = [trace]

layout = go.Layout(
    title='SpaceX Falcon Heavy - Falcon Heavy Test Flight - Stage 1 (0-300 seconds)',
    scene=dict(
        xaxis=dict(
            gridcolor='rgb(0, 0, 0)',
            zerolinecolor='rgb(0, 0, 0)',
            showbackground=False,
            backgroundcolor='rgb(230, 230, 230)'),
        yaxis=dict(
            gridcolor='rgb(0, 0, 0)',
            zerolinecolor='rgb(0, 0, 0)',
            showbackground=False,
            backgroundcolor='rgb(230, 230, 230)'),
        zaxis=dict(
            gridcolor='rgb(0, 0, 0)',
            zerolinecolor='rgb(0, 0, 0)',
            showbackground=False,
            backgroundcolor='rgb(230, 230, 230)'),
        camera=dict(
            up=dict(x=0.1, y=-0.95, z=0.24),
            center=dict(x=0, y=0, z=0),
            eye=dict(x=1.5, y=-0.15, z=-1.15)),
        aspectmode='cube',
        dragmode="orbit"))
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='jupyter-parametric_plot')
```
